## Merge Dataframes & Final Manipulation

```python
import pandas as pd
```

```python
NLC=pd.read_excel("NLC.xlsx")
STK=pd.read_excel("STK.xlsx")
Products_Prices=pd.read_excel("Final_data.xlsx")
```

```python
merged_df = pd.merge(NLC, STK, left_on='MODEL.1', right_on='Model',how='left')
```

```python
merged_df.head(10)
```

```python
NLC.shape
```

```python
images_column = Products_Prices.pop('Images') # remove the column from its current position
Products_Prices.insert(0, 'Images', images_column)
Products_Prices.head()
```

```python
merged_df2=pd.merge(merged_df, Products_Prices, left_on='MODEL', right_on='Model',how='left')
```
